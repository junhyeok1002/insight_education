### dataframe접근 심화
- 기존처럼 하나씩 데이터 프레임에 접근하는 방식 : 오래걸림 <br>
→ 조건을 걸어 그룹단위로 접근

```python
import pandas as pd

lst = [['하정현',18,'경영','빅사','회장'],['나경훈',16,'수학','빅사','부회장'],\
       ['손준석',15,'철학','경영','교육팀장'],['이창윤',19,'경제','빅데이터','교육팀원'],\
       ['신유정',18,'경영','빅사','교육팀원'],['백미정',19,'영문','경영','교육팀원']]

df = pd.DataFrame(lst, columns=['이름','학번','전공','복수전공','역할'])
df = df.set_index('이름')
df
```

## 과제3
과목별 중간고사의 count, sum, max, min, mean, median 정보를 알려주는 데이터프레임을 만들고, <br>
같은 방식으로 기말고사의 데이터프레임을 만든 뒤, 이를 병합하여 아래와 같은 결과를 도출하세요!

```python
# 아래와 같은 결과가 나오면 됩니다!
final = score_df.pivot_table(index = '과목', values = ['기말고사'], \
                     aggfunc = ['count', 'sum', 'max', 'min', 'mean', 'median'])
midterm = score_df.pivot_table(index = '과목', values = ['중간고사'], \
                     aggfunc = ['count', 'sum', 'max', 'min', 'mean', 'median'])
```

```python
display(final)
display(midterm)
```

방법1: final과 midterm를 병합하여 columns을 일일이 맞추어 준다.

```python
score = pd.merge(final,midterm, how = 'outer', left_index = True, right_index = True)
score.groupby('과목')[[('count','기말고사'),('count','중간고사'),
            ('sum','기말고사'),('sum','중간고사'),
            ('max','기말고사'),('max','중간고사'),
            ('min','기말고사'),('min','중간고사'),
            ('mean','기말고사'),('mean','중간고사'),
            ('median','기말고사'),('median','중간고사')]].sum()
#지저분하네요.. 다른방법이 있으면 알려주세요!
```

방법2 : 근데 이건 중간,기말을 병합하는 게 아니지 않나요?

```python
new_score_df = score_df.pivot_table(index = '과목', values= ['기말고사','중간고사'], \
                     aggfunc=['count','sum','max','min','mean','median'])
new_score_df
```
